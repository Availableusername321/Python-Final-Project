```python
# 1. Привести колонку к формату datetime
df['date_new'] = pd.to_datetime(df['date_new'], errors='coerce')

# 2. Теперь можно применять .dt.strftime()
df['date_new_formatted'] = df['date_new'].dt.strftime('%d.%m.%Y')
```

```python
#Exercise 1
def count_unique_users_by_month(df, year, month):
    filtered = df[(df['date_new'].dt.year == year) & (df['date_new'].dt.month == month)]
    return filtered['user_id'].nunique()
```

```python
unique_november_users = count_unique_users_by_month(df, 2023, 11)
print("Уникальных пользователей за ноябрь 2023:", unique_november_users)
```

```python
#Exercise 2
def count_unique_users_by_day(df, year, day):
    filtered = df[(df['date_new'].dt.year == year) & (df['date_new'].dt.day == day)]
    return filtered['user_id'].nunique()
```

```python
unique_daily_active_users = count_unique_users_by_day(df, 2023, 30)
print("Активных уникальных пользователей в день:", unique_daily_active_users)
```

```python
daily_unique_users = df.groupby('date_new')['user_id'].nunique().reset_index()

# Переименуем колонку для ясности
daily_unique_users.rename(columns={'user_id': 'unique_users'}, inplace=True)

print(daily_unique_users)
```

```python
avg_num_users = daily_unique_users.mean()
avg_num_users
```

```python
#Exercise 3
users_nov_1 = df[df['date_new'] == '2023-11-01']['user_id'].unique()

# Пользователи, которые были активны 2 ноября
users_nov_2 = df[df['date_new'] == '2023-11-02']['user_id'].unique()

# Пересечение
retained = set(users_nov_1) & set(users_nov_2)

# Расчёт
retention = len(retained) / len(users_nov_1) if len(users_nov_1) > 0 else 0
print(f"Day 1 retention: {retention:.2%}")
```

```python
#Exercise 5
df['date_new'] = pd.to_datetime(df['date_new'])
november_df = df[(df['date_new'].dt.month == 11) & (df['date_new'].dt.year == 2023)]

# Общее число пользователей в ноябре
total_users = november_df['user_id'].nunique()

# Пользователи, у которых view_adverts >= 1
active_users = november_df[november_df['view_adverts'] > 0]['user_id'].nunique()

# Конверсия
conversion_rate = active_users / total_users

print(f"Пользовательская конверсия в просмотр объявления за ноябрь: {conversion_rate:.2%}")
```

```python
#Exercise 6
num_active_adverts = df[df['view_adverts'] != 0].count()
num_active_adverts
num_adverts = df['view_adverts'].count()
avg_num_adverts = (num_active_adverts/num_adverts)*100
avg_num_adverts
```

```python
#Exercise 6
# Фильтруем только ноябрь 2023
df_november = df[(df['date_new'].dt.month == 11) & (df['date_new'].dt.year == 2023)]

# Группируем по user_id и суммируем просмотры
user_views = df_november.groupby('user_id')['view_adverts'].sum()

# Считаем среднее количество просмотров на пользователя
average_views = user_views.mean()

print(f"Среднее количество просмотренных объявлений на пользователя в ноябре: {average_views:.2f}")
```

```python
#Exercise 7
# Данные опроса
data = {
    'group': ['Promoters', 'Passives', 'Detractors'],
    'count': [1200, 300, 500]
}

df = pd.DataFrame(data)

# Общая сумма опрошенных
total = df['count'].sum()

# Считаем проценты
promoters_pct = df.loc[df['group'] == 'Promoters', 'count'].values[0] / total * 100
detractors_pct = df.loc[df['group'] == 'Detractors', 'count'].values[0] / total * 100

# NPS
nps = promoters_pct - detractors_pct

print(f'NPS = {nps:.0f}')
```

```python
#Exercise 8
import pandas as pd
from scipy.stats import ttest_ind
arpu_summary = ex.groupby('experiment_group')['revenue'].agg(['mean', 'count', 'sum']).rename(columns={'mean': 'ARPU'})
print("ARPU по группам:")
print(arpu_summary)

# Разделение по группам
test_group = ex[ex['experiment_group'] == 'test']['revenue']
control_group = ex[ex['experiment_group'] == 'control']['revenue']

# Welch t-test
t_stat, p_value = ttest_ind(test_group, control_group, equal_var=False)
print(f"\nP-value: {p_value:.4f}")
```

```python
#Exercise 18
conv_A = 1003 / 100047501
conv_B = 1099 / 100001055
print(conv_A)
print(conv_B)
```

```python
from statsmodels.stats.proportion import proportions_ztest

# Кол-во успешных событий (платежи)
successes = [1003, 1099]

# Кол-во наблюдений (посетители)
n_obs = [100047501, 100001055]

z_stat, p_value = proportions_ztest(successes, n_obs)
print(f"Z-статистика: {z_stat:.4f}, p-value: {p_value:.4g}")
```
